```python
def plot_probs(probs, times, joined):
    for i in range(len(probs)):
        y = probs.iloc[i, :]
        y = y[~y.isna()].values

        x = times.iloc[i, :]
        print(len(x))
        x = x[~x.isna()]
        print(len(x))
        x = x[0: len(y)]
        
        truths = joined.loc[idx, :]
        st, et = truths['start_time'], truths['end_time']
        
        fig = px.scatter(x=x, y=y)
        fig.add_shape(
            type='line',
            yref="y",
            xref="x",
            x0=st,
            y0=0,
            x1=st+1,
            y1=max(y),
            line=dict(color='black', width=3)
        )

        fig.add_shape(
            type='line',
            yref="y",
            xref="x",
            x0=et,
            y0=0,
            x1=et+1,
            y1=max(y),
            line=dict(color='black', width=3)
        )
        fig.update_traces(connectgaps=False)
        fig.update_layout(showlegend=False)
        os.makedirs('./inference_results/plots/', exist_ok=True)
        fig.show()
        fig.write_image(f'./inference_results/plots/val_prob_plot_{i}.png')
```

```python
probs = pd.read_csv('inference_results/csv/probs_validation_results_resnet18-unweighted-unbalanced-105k_val.csv', index_col='Unnamed: 0')
probs = probs.rolling(15, min_periods=1).mean()

times = pd.read_csv('inference_results/csv/times_validation_results_resnet18-unweighted-unbalanced-105k_val.csv', index_col='Unnamed: 0')

truth = format_data_csv('src/data/val_na_stratified.csv').set_index('origin_uri')

joined = truth.join(probs, how='inner')

plot_probs(probs, times, joined)
```

```python
THRESH = 0.3 

def thresh(x):
    return int(x > THRESH)

def moving_average(a, n=3):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

def calculate_entry_exit_time(probs, times):
    probs = moving_average(probs, n=15)
    probs = np.mean(np.sort(probs, axis=1)[0: top_10_pct_len])
    
    indices = np.where(probs > THRESH)[0]
    # first and last
    return times[indices[0]], times[indices[-1]]
    
    
    
```
